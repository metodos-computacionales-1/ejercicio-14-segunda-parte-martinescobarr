```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dat = np.loadtxt('rk4sinfriccion.dat')
t=dat[:,0]
x=dat[:,1]
v=dat[:,2]
```

```python
plt.figure()
plt.plot(t,x,label='Posición')
plt.xlabel('t')
plt.ylabel('x')
plt.title("Posición")
plt.legend()
plt.savefig('pos.png')

plt.figure()
plt.plot(x,v,label="x vs v")
plt.xlabel('x')
plt.ylabel('v')
plt.title("velocidad")
plt.legend()
plt.savefig('vel.png')
```
